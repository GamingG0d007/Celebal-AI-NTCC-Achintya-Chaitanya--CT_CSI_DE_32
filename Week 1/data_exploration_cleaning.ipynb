# Python Basics - Data Exploration and Cleaning with Pandas
## Clothing Set Dataset Analysis

```python
import pandas as pd
import numpy as np

# Step 1: Load the CSV dataset into a Pandas DataFrame
file_path = 'archive/clothing-set.csv'
df = pd.read_csv(file_path)
print(f'Dataset loaded with shape: {df.shape}')
```

```python
# Step 2: Explore data
print('=== HEAD ===')
print(df.head())
print('\n=== TAIL ===')
print(df.tail())
print('\n=== SHAPE ===')
print(df.shape)
print('\n=== COLUMNS ===')
print(df.columns.tolist())
print('\n=== DATA TYPES ===')
print(df.dtypes)
print('\n=== SUMMARY ===')
print(df.describe())
```

```python
# Step 3: Handle missing values
print('=== MISSING VALUES ===')
print(df.isnull().sum())

# Fill or drop missing values
# For numeric columns, fill with median
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df[col].isnull().any():
        df[col].fillna(df[col].median(), inplace=True)

# For categorical columns, fill with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().any():
        df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'Unknown', inplace=True)

print('\n=== MISSING VALUES AFTER CLEANING ===')
print(df.isnull().sum())
```

```python
# Step 4: Perform basic operations
# Filter rows - e.g., products with rating >= 4.0
filtered_df = df[df['rating'] >= 4.0]
print(f'Filtered rows (rating >= 4.0): {len(filtered_df)}')

# Select specific columns
selected_df = df[['product_id', 'title', 'initial_price', 'final_price', 'rating']]
print('\n=== SELECTED COLUMNS ===')
print(selected_df.head())
```

```python
# Step 5: Remove duplicates
print(f'Rows before removing duplicates: {len(df)}')
df.drop_duplicates(inplace=True)
print(f'Rows after removing duplicates: {len(df)}')
```

```python
# Step 6: Create a derived column (total_amount = final_price * 1, assuming quantity=1 for simplicity)
# Note: The dataset doesn't have a quantity column, so we'll use final_price as total_amount
df['total_amount'] = df['final_price']

# Alternatively, if we want to calculate based on initial_price and discount
df['total_amount_calculated'] = df['initial_price'] * (1 - df['discount'] / 100)

print('\n=== DERIVED COLUMNS ===')
print(df[['product_id', 'initial_price', 'discount', 'final_price', 'total_amount', 'total_amount_calculated']].head())
```

```python
# Step 7: Save the cleaned dataset as a new CSV file
output_path = 'cleaned_clothing_set.csv'
df.to_csv(output_path, index=False)
print(f'Cleaned dataset saved to: {output_path}')
print(f'Final shape: {df.shape}')
```
